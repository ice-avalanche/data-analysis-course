## 1. What is an API?

- **Definition**:
  - **API (Application Programming Interface)**: 
    - A set of rules, protocols, and tools that allow different software applications to communicate with each other.
    - It defines how software components should interact, making it easier to integrate and extend functionality.
  - **Purpose**:
    - APIs enable developers to access the functionality of another application or service without needing to understand its internal workings.
    - They are commonly used to retrieve data, perform actions, or integrate services.

- **Types of APIs**:
  1. **REST (Representational State Transfer)**:
     - A widely used architectural style for designing networked applications.
     - Uses standard HTTP methods (GET, POST, PUT, DELETE) to perform operations.
     - Data is typically exchanged in JSON or XML format.
     - Example: Fetching stock prices from a financial API.
  2. **SOAP (Simple Object Access Protocol)**:
     - A protocol for exchanging structured information in web services.
     - Uses XML for message formatting and relies on HTTP or SMTP for communication.
     - More rigid and complex than REST, but offers built-in error handling and security features.
     - Example: Enterprise-level web services.

- **Why APIs Matter**:
  - APIs allow developers to build applications faster by reusing existing services.
  - They enable data sharing between systems, which is crucial for modern applications.
  - APIs power integrations between platforms, such as connecting a website to a payment gateway or fetching weather data for a mobile app.

## 2. How APIs Work

- **Request-Response Model**:
  - APIs operate on a request-response model:
    1. **Request**: The client (e.g., your Python script) sends a request to the API server.
    2. **Response**: The server processes the request and sends back a response, usually in JSON or XML format.
  - Example:
    - Request: `GET https://api.example.com/data?param=value`
    - Response: `{"key": "value"}`

- **HTTP Methods**:
  - **GET**: Retrieve data from the server (e.g., fetch stock prices).
  - **POST**: Send data to the server to create a new resource (e.g., submit a form).
  - **PUT**: Update an existing resource on the server (e.g., edit a user profile).
  - **DELETE**: Remove a resource from the server (e.g., delete a record).

- **Endpoints**:
  - A specific URL where an API can be accessed.
  - Example: `https://api.example.com/users` (to access user data).

## 3. Practical Example: Using an API in Python

This Python script demonstrates how to interact with a financial API (specifically, Alpha Vantage) to fetch daily stock data for a given symbol (in this case, Microsoft, represented by `MSFT`). Below is a step-by-step breakdown of the code:

#### 3.1. Importing the `requests` Library

````python
import requests
````

- The `requests` library is used to send HTTP requests in Python.
- It simplifies making API calls and handling responses.

#### 3.2. Defining the API Endpoint and Parameters

````python
url = 'https://www.alphavantage.co/query'
params = {
    'function': 'TIME_SERIES_DAILY',
    'symbol': 'MSFT',
    'apikey': 'your_api_key'
}
````

#### 3.3. Sending a GET Request

````python
response = requests.get(url, params=params)
````

- **`requests.get()`**: Sends a GET request to the API endpoint.
- **`url`**: The API endpoint URL.
- **`params`**: The query parameters passed to the API.
- The response from the API is stored in the `response` variable.

#### 3.4. Checking if the Request Was Successful

````python
if response.status_code == 200:
    pass # here should be code up to this condition
````

- **`response.status_code`**: The HTTP status code returned by the server.
  - **`200`**: Indicates a successful request.
  - Other common status codes:
    - **`400`**: Bad request (e.g., invalid parameters).
    - **`401`**: Unauthorized (e.g., missing or invalid API key).
    - **`404`**: Not found (e.g., invalid endpoint).
    - **`500`**: Internal server error.

#### 3.5. Parsing the JSON Response

````python
if response.status_code == 200:
    data = response.json()  # Parse JSON response
    print(data)
````

- **`response.json()`**: Converts the JSON response from the API into a Python dictionary.
- The parsed data is stored in the `data` variable.
- **`print(data)`**: Displays the fetched data in the console.

#### 3.6. Handling Errors

````python
if response.status_code == 200:
    data = response.json()  # Parse JSON response
    print(data)
else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")
````

- If the request fails (e.g., status code is not `200`), an error message is printed.
- The status code is included in the message to help debug the issue.

````python
# full code

import requests
import json

# API endpoint and parameters
url = 'https://www.alphavantage.co/query'
params = {
    'function': 'TIME_SERIES_DAILY',
    'symbol': 'MSFT',
    'apikey': 'your_api_key'
}

# Sending a GET request
response = requests.get(url, params=params)

# Checking if the request was successful
if response.status_code == 200:
    data = response.json()  # Parse JSON response
    print(json.dumps(data, indent=4, ensure_ascii=False))
else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")
````

## 4. HTTP Methods

- **Definition**:
  - HTTP (Hypertext Transfer Protocol) methods are actions that clients (e.g., browsers, apps) can perform on resources hosted on a server.
  - They are the foundation of communication between clients and servers in web applications and APIs.
- **Purpose**:
  - HTTP methods define the type of action to be performed on a resource, such as retrieving data, creating new records, updating existing ones, or deleting them.

### 4.1. GET

   - **Purpose**: Retrieve data from the server.
   - **Use Case**: Fetching information, such as stock prices, user profiles, or weather data.
   - **Example**:

````python
import requests
import json

url = 'https://api.openweathermap.org/data/2.5/weather'
params = {
    'q': 'Saint Petersburg,ru',
    'appid': 'your_api_key',
    'units': 'metric',
    'lang': 'en'
}

try:
    response = requests.get(url, params=params)

    if response.status_code == 200:
        data = response.json()
        print("Success!")
        print("\nSt. Petersburg Data:")
        print(json.dumps(data, indent=4, ensure_ascii=False))
    else:
        print(f"Response error: {response.status_code}")
        
except requests.exceptions.RequestException as e:
    print(f"Request Error: {e}") 
````

### 4.2. POST

   - **Purpose**: Send data to the server to create a new resource.
   - **Use Case**: Submitting forms, creating new records (e.g., adding a new user to a database).
   - **Example**:

````python
import requests
import json

# Using JSONPlaceholder API for testing POST request
url = 'https://jsonplaceholder.typicode.com/posts'

# Data to send
data = {
    'title': 'Test Title',
    'body': 'This is a test message in English',
    'userId': 1
}

try:
    # Making POST request
    response = requests.post(url, json=data)
    
    # Checking response status
    if response.status_code == 201:  # 201 means successful creation
        # Getting JSON data
        result = response.json()
        print("POST request successful!")
        print("\nCreated data:")
        # Using json.dumps with indent parameter for pretty printing
        print(json.dumps(result, indent=4, ensure_ascii=False))
    else:
        print(f"Error! Status code: {response.status_code}")
        
except requests.exceptions.RequestException as e:
    print(f"Request error occurred: {e}") 
````

### 4.3. PUT

   - **Purpose**: Update an existing resource on the server.
   - **Use Case**: Modifying data, such as updating a user's profile or changing an order's status.
   - **Example**:

````python
import requests
import json

# Using JSONPlaceholder API for testing PUT request
url = 'https://jsonplaceholder.typicode.com/posts/1'  # ID of the post to update

# Updated data
data = {
    'id': 1,
    'title': 'Updated Title',
    'body': 'This is an updated message',
    'userId': 1
}

try:
    # Making PUT request
    response = requests.put(url, json=data)
    
    # Checking response status
    if response.status_code == 200:  # 200 means successful update
        # Getting JSON data
        result = response.json()
        print("PUT request successful!")
        print("\nUpdated data:")
        # Using json.dumps with indent parameter for pretty printing
        print(json.dumps(result, indent=4, ensure_ascii=False))
    else:
        print(f"Error! Status code: {response.status_code}")
        
except requests.exceptions.RequestException as e:
    print(f"Request error occurred: {e}") 
````

### 4.4. DELETE

   - **Purpose**: Remove a resource from the server.
   - **Use Case**: Deleting records, such as removing a user or canceling an order.
   - **Example**:

````python
import requests

# Using JSONPlaceholder API for testing DELETE request
url = 'https://jsonplaceholder.typicode.com/posts/1'  # ID of the post to delete

try:
    # Making DELETE request
    response = requests.delete(url)
    
    # Checking response status
    if response.status_code == 200:  # JSONPlaceholder returns 200 for successful deletion
        print("DELETE request successful!")
        print("Resource has been deleted")
    else:
        print(f"Error! Status code: {response.status_code}")
        
except requests.exceptions.RequestException as e:
    print(f"Request error occurred: {e}") 
````

This can be useful for gathering data that is not provided through an API or for automating repetitive tasks such as price monitoring, news aggregation, or competitor analysis.

**Beautiful Soup** is a popular Python library that simplifies the parsing of HTML and XML documents.
````python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

def parse_currency_data():
    # URL of the currency data from CBR
    url = 'https://www.cbr.ru/currency_base/daily/'
    
    try:
        # Making GET request
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
        }
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        
        # Creating BeautifulSoup object
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Finding the table with currency data
        table = soup.find('table', {'class': 'data'})
        
        if table:
            # List to store currency data
            currency_data = []
            
            # Getting all rows except header
            rows = table.find_all('tr')[1:]
            
            # Current date
            current_date = datetime.now().strftime('%Y-%m-%d')
            
            # Processing each row
            for row in rows:
                cols = row.find_all('td')
                if len(cols) >= 5:
                    currency_info = {
                        'date': current_date,
                        'code': cols[1].text.strip(),
                        'name': cols[2].text.strip(),
                        'nominal': cols[3].text.strip(),
                        'rate': float(cols[4].text.strip().replace(',', '.'))
                    }
                    currency_data.append(currency_info)
            
            # Converting to DataFrame
            df = pd.DataFrame(currency_data)
            
            # Saving to CSV
            df.to_csv('currency_data.csv', mode='a', header=not pd.io.common.file_exists('currency_data.csv'), index=False)
            
            return df
            
        else:
            print("Currency data table not found")
            return None
            
    except requests.exceptions.RequestException as e:
        print(f"Error fetching data: {e}")
        return None
    except Exception as e:
        print(f"Error parsing data: {e}")
        return None

def analyze_currency_data(df):
    if df is None:
        return
    
    print("\n=== Basic Statistics ===")
    print("\nTop 5 currencies by rate:")
    print(df.nlargest(5, 'rate')[['code', 'name', 'rate']])
    
    print("\nBottom 5 currencies by rate:")
    print(df.nsmallest(5, 'rate')[['code', 'name', 'rate']])
    
    print("\nSummary statistics for rates:")
    print(df['rate'].describe())
    
    # Creating visualizations
    plt.figure(figsize=(12, 6))
    
    # Bar plot for top 10 currencies
    plt.subplot(1, 2, 1)
    top_10 = df.nlargest(10, 'rate')
    sns.barplot(data=top_10, x='code', y='rate')
    plt.title('Top 10 Currencies by Rate')
    plt.xticks(rotation=45)
    
    # Distribution plot
    plt.subplot(1, 2, 2)
    sns.histplot(data=df, x='rate', bins=20)
    plt.title('Distribution of Currency Rates')
    
    plt.tight_layout()
    plt.savefig('currency_analysis.png')
    print("\nVisualization saved as 'currency_analysis.png'")

def main():
    print("Fetching currency data...")
    df = parse_currency_data()
    
    if df is not None:
        print("\nData successfully parsed!")
        print("\nSample of parsed data:")
        print(df.head())
        
        print("\nStarting analysis...")
        analyze_currency_data(df)
    else:
        print("Failed to analyze data due to parsing error")

main()
````
